```python
import pandas as pd
df = pd.read_csv('laptops.csv',sep=',',encoding ='cp1251')
df.head()
```

## Задание 1
### 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU (Intel, AMD и т.д)    

```python
#df['Cpu_productor']=df['Cpu'].str.split(pat=' ',n=0)
def my_split(s:str) -> str:
    return s.split(' ')[0]
df['Cpu_productor']=df['Cpu'].apply(my_split)
df.head()
```

### 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя (128, 1024, 256 и т.д.)

```python
def my_mem(s:str) -> int:
    if 'TB' in s.split(' ')[0]:
        return int(float(s.split(' ')[0].strip().replace('TB',''))*1024)
    else: return int(float(s.split(' ')[0].strip().replace('GB','')))
    #return s.split(' ')[0].strip().strip().replace('GB','')
df['Memory_Amount']=df['Memory'].apply(my_mem)
df.head()
```

### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD, SDD и т.д.)    

```python
df['Memory'].value_counts()
```

```python
def my_mem_type(s:str) -> str:
    return s.split('+')[0].split(' ',1)[1].strip()

df['Memory_Type']=df['Memory'].apply(my_mem_type)
df.head()
```

```python
df['Memory_Type'].value_counts()
```

### 1.4 Удалите признаки Memory и ScreenResolution 

```python
df.drop(columns = ['Memory', 'ScreenResolution'], inplace = True)
df.head()
```

## Задание 4.                    
Составьте несколько сводных таблиц                
### 4.1 Найдите среднюю стоимость ноутбуков в зависимости от компании производителя        
Отсортируйте от меньшей стоимости к большей        

```python
df.groupby(['Company']).mean()['Price_euros'].sort_values()
```

### 4.2 Найдите минимальную, среднюю и максимальную стоимости ноутбуков в зависимости от производителя процессора    

```python
df.groupby(['Cpu_productor']).agg({'Price_euros':['min','mean', 'max']})
```

### 4.3 Постройте таблицу с подсчетом количества ноутбуков в данных в зависимости от производителя CPU и ОЗУ    

```python
df.pivot_table(index = 'Cpu_productor', columns = 'Ram', values = 'TypeName', aggfunc = 'count')
```

### 4.4 Постройте таблицу с подсчетом средней стоимости ноутбуков в данных в зависимости от операционной системы и GB памяти 

```python
df.pivot_table(index = 'OpSys', columns = 'Memory_Amount', values = 'Price_euros', aggfunc = 'mean')
```
